```python
class CNN(nn.Module):
    iteration =0
    lossA =[]
    accA =[]
    def __init__(self):
        super(CNN, self).__init__()
        self.iteration =0
        self.lossA =[]
        self.accA =[]
        self.conv1 = nn.Sequential(         
            nn.Conv2d(
                in_channels=3,              
                out_channels=16,            
                kernel_size=5,              
                stride=1,                   
                padding=2,                  
            ),                              
            nn.ReLU(),                      
            nn.MaxPool2d(kernel_size=2),    
        )
        self.conv2 = nn.Sequential(         
            nn.Conv2d(16, 64, 5, 1, 2),     
            nn.ReLU(),                      
            nn.MaxPool2d(2),                
        )
        # fully connected layer, output 2 classes
        self.out = nn.Linear(1960000, 2)
    def forward(self, x):
        x = self.conv1(x)
        x = self.conv2(x)        
        # flatten the output of conv2 to (batch_size, 32 * 7 * 7)
        x = x.view(x.size(0), -1)  
        
        output = self.out(x)
        return output, x    # return x for visualization
    def accurancy (self,output, b_y):
        pred_y = torch.max(output, 1)[1].data.squeeze()
        acc = (pred_y == b_y).sum().item() / float(b_y.size(0))
        return acc*100

    def plot(self,iteration,loss,acc):
        iteration = range(0,iteration)
        plt.plot(iteration, loss, 'g', label='Training loss')
        
    def validation(self):
        with torch.no_grad():
            for images, labels in loaders['test']:
                test_output, last_layer = self(images)
                pred_y = torch.max(test_output, 1)[1].data.squeeze()
                accuracy = (pred_y == labels).sum().item() / float(labels.size(0))
                pass
        print('Test Accuracy of the model on the 10000 test images: %.2f' % accuracy*100)

    def pred(self, images):
        with torch.no_grad():
            test_output, last_layer = self(images)
            print ("from model",test_output, last_layer)
            print("")
            return(test_output)

    def train(self,num_epochs,loaders):    
        total_step = len(loaders)        
        for epoch in range(num_epochs):
            for i, (images, labels) in enumerate(loaders):
                self.iteration = self.iteration+1
                # gives batch data, normalize x when iterate train_loader
                b_x = Variable(images)   # batch x
                b_y = (labels)   # batch y
                output = self(b_x)[0]               
                loss = loss_func(output, b_y)
                acc =self.accurancy(output, b_y)   
                self.lossA.append(loss.item())
                self.accA.append(acc)
                # clear gradients for this training step   
                optimizer.zero_grad()           

                # backpropagation, compute gradients 
                loss.backward()    
                # apply gradients             
                optimizer.step()                

                if (i+1) % 2 == 0:
                    print ('Epoch [{}/{}], Step [{}/{}], Loss: {:.4f}, Accurancy : {:.1f}'.format(epoch + 1, num_epochs, i + 1, total_step, loss.item(),acc))
            self.plot(self.iteration,self.lossA,self.accA)
            
```

```python
cnn = CNN()
print(cnn)
loss_func = nn.CrossEntropyLoss()   
loss_func
optimizer = optim.Adam(cnn.parameters(), lr = 0.001)  
num_epochs = 10 
cnn.train(num_epochs, loaders)

```
